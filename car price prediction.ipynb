### Log transformation

```python
data_log = pd.DataFrame([])
data_log['Price'] = np.log(df5['Price'])
data_log['EngineV'] = np.log(df5['EngineV'])
```

'Price' and 'Mileage' were the variables with left skewed

### Feature scaling and normalization

```python
mile = x['Mileage']
logengv = data_log['EngineV']

scaled_mile = (mile - min(mile)) / (max(mile) - min(mile))
scaled_engv = (logengv - np.mean(logengv)) / np.sqrt(np.var(logengv))

data_cont_scaled = pd.DataFrame([])
data_cont_scaled['Mileage'] = scaled_mile
data_cont_scaled['EngineV'] = scaled_engv
```

### Putting the final data together

```python
df_fin = pd.DataFrame([])
df_fin['Mileage'] = scaled_mile
df_fin['EngineV'] = scaled_engv
df_fin['Year'] = df5['Year']
Brand_Dummies = pd.get_dummies(df5['Brand'], prefix = 'Brd', drop_first = True)
Body_Dummies = pd.get_dummies(df5['Body'], prefix = 'Bdy', drop_first = True)
Engine_Type_Dummies = pd.get_dummies(df5['Engine Type'], prefix = 'EngT', drop_first = True)
Registration_Dummies = pd.get_dummies(df5['Registration'], prefix = 'Rego', drop_first = True)
Price = np.log(df5['Price'])
df_fin = pd.concat([Price, df_fin, Brand_Dummies, Body_Dummies, Engine_Type_Dummies, Registration_Dummies], axis=1)
```

## Model

This section will model the final data using the OLS regression model

```python
outcome = 'Price'
predictors = df_fin.drop('Price', axis = 1)
pred_sum = '+'.join(predictors.columns)
formula = outcome + '~' + pred_sum
model = ols(formula=formula, data=df_fin).fit()
print(model.summary())
```

### Model Validation

```python
x1 = df_fin.drop('Price', axis=1)
y1 = df_fin['Price']

x1_train, x1_test, y1_train, y1_test = train_test_split(x1, y1, test_size=0.2, random_state=42)

linreg = LinearRegression()
linreg.fit(x1_train, y1_train)

y1_hat_train = linreg.predict(x1_train)
y1_hat_test = linreg.predict(x1_test)

train_residuals = y1_hat_train - y1_train
test_residuals = y1_hat_test - y1_test

mse_train = np.sum((y1_train-y1_hat_train)**2)/len(y1_train)
mse_test = np.sum((y1_test-y1_hat_test)**2)/len(y1_test)
print('Train Mean Squarred Error:', mse_train)
print('Test Mean Squarred Error:', mse_test)
```
